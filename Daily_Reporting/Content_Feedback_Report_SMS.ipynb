```python
def parse_pieces(row):
    if row is np.nan:
        return row
    else:
        try:
            return int(row)
        except:
            row = re.sub("[a-zA-Z &,!.]", "", row)
            row = row.split('-')[-1]
            if (len(row) <=2) & (len(row) > 0):
                return int(row)
            else:
                return np.nan 
```
